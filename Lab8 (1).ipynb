```python
import os


base_dir = "lab_data"
if not os.path.exists(base_dir):
    os.makedirs(base_dir)
    print("Directory ",base_dir," created.")

subfolders = ["input", "output", "logs"]
for folder in subfolders:
    path = os.path.join(base_dir, folder)
    if not os.path.exists(path):
        os.makedirs(path)

print("\nItems in lab_data:")
for item in os.listdir(base_dir):
    item_path = os.path.join(base_dir, item)
    type_info = "Directory" if os.path.isdir(item_path) else "File"
    print(item," : ",type_info)

old_log_path = os.path.join(base_dir, "logs")
new_log_path = os.path.join(base_dir, "history")
if os.path.exists(old_log_path):
    os.rename(old_log_path, new_log_path)
    print("\nRenamed 'logs' to 'history'.")
```

```python
import os

notes_path = os.path.join(base_dir, "input", "notes.txt")
lines = ["Line 1: Hello World\n", "Line 2: Python Lab\n", 
         "Line 3: File Handling\n", "Line 4: OS Module\n", "Line 5: Final Line\n"]

with open(notes_path, "w") as f:
    f.writelines(lines)


with open(notes_path, "r") as f:
    content = f.readlines()
    print("\nTotal lines in notes.txt: ",len(content))

import shutil
summary_path = os.path.join(base_dir, "output", "summary.txt")
shutil.copy(notes_path, summary_path)
print("Content copied to summary.txt.")
```

```python
import os
import pyfiglet 
ascii_art = pyfiglet.figlet_format("Dahy")
ascii_file_path = os.path.join(base_dir, "output", "ascii_art.txt")

with open(ascii_file_path, "w") as f:
    f.write(ascii_art)
print("\nASCII art saved to output/ascii_art.txt.")
```
